## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['line_subtotal'] = df['unit_price'] * df['qty']
df[['unit_price', 'qty', 'line_subtotal']].head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['total_weight'].apply(lambda weight: weight * 7 if weight > 50 else weight * 10)
df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['line_price'] = ((df['line_subtotal'] + df['shipping_price'])*1.0925).round(2)
df[['line_subtotal', 'shipping_price', 'line_price']].head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']
df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
df['profit'] = df['line_price'] - df['line_cost']
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
for client in top_5_clients:
    print(f'{client}: ${df.loc[df["client_id"]==client, "line_price"].sum():.2f}')
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
top_5_df = df[df['client_id'].isin(top_5_clients)]
top_5_summary = top_5_df.groupby('client_id')[['qty', 'shipping_price', 'line_price', 'profit']].sum()
top_5_summary = top_5_summary.reset_index()
top_5_summary
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 
money_cols = ['shipping_price', 'line_price', 'profit']

# Define a function that converts a dollar amount to millions.
def as_millions(value): 
    """
    Divides a number by 1,000,000 and rounds the result to 3 decimal places
    Args:
        value: a float or integer
    Returns:
        the value argument divided by 1,000,000
    Raises:
        None
    """
    return value/1_000_000

# Apply the currency_format_millions function to only the money columns. 
top_5_summary[money_cols] = top_5_summary[money_cols].apply(as_millions, axis=1)


# Rename the columns to reflect the change in the money format. 
top_5_summary.rename(columns={'client_id': 'Client ID',
                              'qty':'Units',
                              'shipping_price':'Shipping (millions)',
                              'line_price':'Total Revenue (millions)',
                              'profit':'Total Profit (millions)'}, inplace=True)

top_5_summary
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
top_5_summary_sorted = top_5_summary.sort_values(by='Total Profit (millions)', ascending=False)
top_5_summary_sorted
```

## Findings  
While the 'top_5_summary_sorted' dataframe shows the top 5 clients based on *how many records they have in the dataset*, it's worth noting that only one of these clients were among the top 5 based on *overall profitability*. Customers having a large quantity of separate orders does not necessarily translate to bringing the most profit to the company. That may not have been feasible to discover from the dataset without first performing a series of data manipulation and feature engineering methods via pandas. This goes to show that good EDA techniques are necessary to truly understand the data before drawing conclusions.

```python
# Applying the same grouping, summing, and sorting steps but to the original dataset
top_5_by_profit = df.groupby('client_id')[['qty', 'shipping_price', 'line_price', 'profit']].sum()
top_5_by_profit = top_5_by_profit.sort_values(by='profit', ascending=False).reset_index()
top_5_by_profit[money_cols] = top_5_by_profit[money_cols].apply(lambda row: as_millions(row), axis=0)

# Renaming the columns to reflect the change in the money format. 
top_5_by_profit.rename(columns={'client_id': 'Client ID',
                              'qty':'Units',
                              'shipping_price':'Shipping (millions)',
                              'line_price':'Total Revenue (millions)',
                              'profit':'Total Profit (millions)'}, inplace=True)

# Displaying the top 5 customers by profit
top_5_by_profit.head()
```
